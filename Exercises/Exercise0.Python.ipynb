````python
# Extraherar alla löner
salaries = [employee['salary'] for employee in employees]

# Summerar alla löner
total_salary = sum(salaries)

# Dela summan med antalet anställda för att få genomsnittet
average_salary = total_salary / len(employees)

print(f"The average salary in Hire IT is {average_salary:.0f}kr.")
````

## Theory questions



### a) What is the difference between dictionary and lists?

Listor och ordböcker är två olika sätt att lagra data i Python.

- **Listor (Lists)**: 
  En lista är som en rad lådor där varje låda har en platsnummer. Du kan lägga vad du vill i varje låda och använda platsnumret för att hämta det. Listor är ordnade, vilket betyder att ordningen du lägger till saker är viktig och du kan alltid få tillbaka dem i den ordningen. Du kan även ändra innehållet i en lista efter att du skapat den.

  - Exempel:
    ```python
    min_lista = [1, 2, 3, 'äpple']
    print(min_lista[0])  # Output: 1 (Första elementet i listan)
    ```

- **Ordböcker (Dictionaries)**:
  En ordbok är som en uppslagsbok där varje "ord" (nyckel) är kopplat till en "definition" (värde). Istället för att använda platsnummer använder du nyckeln för att hitta det du söker. Ordböcker är inte ordnade, så det spelar ingen roll i vilken ordning du lägger till saker. Du kan också ändra innehållet i en ordbok.

  - Exempel:
    ```python
    min_ordbok = {'namn': 'Alice', 'ålder': 30}
    print(min_ordbok['namn'])  # Output: Alice (Hämtar värdet kopplat till nyckeln 'namn')
    ```

### b) What are positional arguments in functions?

Positionsargument är som att följa ett recept där du lägger till ingredienser i en specifik ordning. När du anropar en funktion, måste du ge argumenten i den ordning som funktionen förväntar sig. Detta gör att Python kan matcha varje argument med rätt parameter i funktionen.

  - Exempel:
    ```python
    def hälsa(namn, ålder):
        print(f"Hej {namn}, du är {ålder} år gammal.")
    
    hälsa('Alice', 30)  # 'Alice' är positionsargument för namn, och 30 för ålder
    ```

### c) What is the difference between using positional arguments and keyword arguments in functions?

- **Positionsargument**: Här ger du argumenten i en specifik ordning. Om funktionen förväntar sig två argument i ordningen namn och ålder, måste du ange dem i just den ordningen.

- **Nyckelargument (Keyword arguments)**: Här kan du ange vad varje argument är genom att använda namn. Det betyder att du inte behöver följa den specifika ordningen och det gör din kod lättare att läsa och förstå.

  - Exempel:
    ```python
    def hälsa(namn, ålder):
        print(f"Hej {namn}, du är {ålder} år gammal.")
    
    hälsa(ålder=30, namn='Alice')  # Argumenten är angivna med nycklar, så ordningen spelar ingen roll
    ```

### d) What is the difference between a list and a tuple?

- **Listor (Lists)**:
  Listor är som en förändringsbar lista över saker. Du kan lägga till, ta bort eller ändra element i en lista efter att den har skapats. Detta gör listor mycket flexibla och användbara när du behöver kunna ändra innehållet.

  - Exempel:
    ```python
    min_lista = [1, 2, 3]
    min_lista[0] = 10  # Ändrar första elementet i listan till 10
    ```

- **Tupler (Tuples)**:
  Tupler är som en låda där du sätter in saker och sedan inte kan ändra dem. När du har lagt något i en tuple, kan du inte ändra, lägga till eller ta bort något. Detta gör tupler användbara när du vill säkerställa att data förblir konstant.

  - Exempel:
    ```python
    min_tuple = (1, 2, 3)
    # min_tuple[0] = 10  # Detta skulle ge ett fel eftersom tupler är oföränderliga
    ```

### e) Explain the concept of Python's dynamic typing.

- **Dynamisk typning** innebär att du inte behöver specificera vilken typ av data en variabel ska vara när du skapar den. Python bestämmer automatiskt datatypen baserat på vad du tilldelar variabeln. Du kan ändra datatypen när som helst utan att få fel.

    ```python
    x = 10  # x är en integer
    ```

### f) What is the purpose behind the `__repr__()` method in a class?

- **`__repr__()`**-metoden är som att ge en detaljerad och teknisk beskrivning av ett objekt. Syftet är att ge en representation som är användbar för utvecklare och ofta kan användas för att återskapa objektet. Det är en exakt och teknisk beskrivning.

  - Exempel:
    ```python
    class Person:
        def __init__(self, namn, ålder):
            self.namn = namn
            self.ålder = ålder
        
        def __repr__(self):
            return f"Person(namn='{self.namn}', ålder={self.ålder})"
    
    p = Person('Alice', 30)
    print(repr(p))  # Output: Person(namn='Alice', ålder=30)
    ```

### g) How does `__str__()` method differ from `__repr__()` method?

- **`__str__()`**-metoden ger en vänlig och lättläst beskrivning av objektet, avsedd för användare. Den är mer användarvänlig och presenterar objektet på ett sätt som är lätt att förstå.

- **`__repr__()`**-metoden ger en teknisk och detaljerad beskrivning, avsedd för utvecklare. Den ska ge en exakt representation som ofta kan användas för att återskapa objektet.

  - Exempel:
    ```python
    class Person:
        def __init__(self, namn, ålder):
            self.namn = namn
            self.ålder = ålder
        
        def __str__(self):
            return f"{self.namn} är {self.ålder} år gammal."
        
        def __repr__(self):
            return f"Person(namn='{self.namn}', ålder={self.ålder})"
    
    p = Person('Alice', 30)
    print(str(p))   # Output: Alice är 30 år gammal.
    print(repr(p))  # Output: Person(namn='Alice', ålder=30)
    ```

### h) Why do you want to encapsulate objects?

- **Inkapsling** är som att lägga något i en skyddande låda för att hålla det säkert och enkelt att använda. Det gör att du kan skydda data och funktioner från att bli ändrade oavsiktligt. Genom att endast erbjuda en kontrollerad väg att interagera med objektet, gör det koden mer pålitlig och lättare att hantera.

### i) What are the pillars of OOP?

De grundläggande pelarna i objektorienterad programmering (OOP) är:

1. **Inkapsling (Encapsulation)**:
   - Skyddar objektets data genom att dölja detaljer och endast ge tillgång till det som är nödvändigt via specifika metoder. Detta gör objektet säkrare och enklare att använda.

2. **Arv (Inheritance)**:
   - Gör det möjligt att skapa nya objekt baserade på befintliga objekt. Det låter dig återanvända kod och bygga vidare på tidigare skapade objekt utan att duplicera kod.

3. **Polymorfism (Polymorphism)**:
   - Tillåter att använda samma metodnamn för olika typer av objekt. Python kommer att använda den version av metoden som är relevant för det aktuella objektet, vilket gör koden mer flexibel och lättare att hantera.

4. **Abstraktion (Abstraction)**:
   - Förenklar komplexa system genom att dölja detaljer och endast visa de viktigaste funktionerna som användaren behöver interagera med. Detta gör det lättare att använda och förstå objektet utan att behöva bry sig om alla interna detaljer.

### Glossory

Här är en detaljerad förklaring av varje term, med tydliga exempel:

---

### Dictionary

En ordbok är en samling av nyckel-värde-par. Varje nyckel är unik och används för att hämta det tillhörande värdet.

- **Exempel**:
  ```python
  my_dict = {'name': 'Alice', 'age': 30}
  print(my_dict['name'])  # Output: Alice
  ```

  Här har vi en ordbok där nyckeln `'name'` är kopplad till värdet `'Alice'`, och nyckeln `'age'` är kopplad till värdet `30`.

---

### List

En lista är en ordnad samling av element som kan ändras. Du kan lägga till, ta bort och ändra element i en lista.

- **Exempel**:
  ```python
  my_list = [1, 2, 3, 'apple']
  my_list.append('banana')
  print(my_list)  # Output: [1, 2, 3, 'apple', 'banana']
  ```

  Här skapar vi en lista med elementen `1, 2, 3` och `'apple'`. Sedan lägger vi till `'banana'` till slutet av listan.

---

### Record

En record är en rad med data som representerar ett objekt eller en entitet. Varje record har flera fält.

- **Exempel**:
  Om vi tänker oss en tabell med personer, kan varje rad (record) se ut så här:
  ```
  ID | Name  | Age
  1  | Alice | 30
  2  | Bob   | 25
  ```

  Här är varje rad en record som innehåller fälten `ID`, `Name` och `Age`.

---

### List Comprehension

List comprehension är en kortfattad syntax för att skapa listor baserat på befintliga listor eller itererbara objekt.

- **Exempel**:
  ```python
  squares = [x**2 for x in range(5)]
  print(squares)  # Output: [0, 1, 4, 9, 16]
  ```

  Här skapar vi en lista med kvadraterna av talen från 0 till 4.

---

### Append

`append` är en metod som lägger till ett element i slutet av en lista.

- **Exempel**:
  ```python
  my_list = [1, 2, 3]
  my_list.append(4)
  print(my_list)  # Output: [1, 2, 3, 4]
  ```

  Här lägger vi till `4` till slutet av listan `my_list`.

---

### Field

Ett fält är en del av en record och representerar en specifik typ av information.

- **Exempel**:
  I en personrecord kan fälten vara:
  ```
  Name: Alice
  Age: 30
  Address: 123 Street
  ```

  Här är `Name`, `Age` och `Address` fält som beskriver en person.

---

### Row

En rad i en tabell eller databas representerar en enskild record med flera fält.

- **Exempel**:
  ```
  Name   | Age | Address
  Alice  | 30  | 123 Street
  Bob    | 25  | 456 Avenue
  ```

  Här representerar varje rad en enskild person med deras namn, ålder och adress.

---

### Column

En kolumn i en tabell eller databas representerar ett specifikt fält som är gemensamt för alla rader.

- **Exempel**:
  ```
  Name
  Alice
  Bob
  ```

  Här är kolumnen `Name` som visar namnen på alla personer i tabellen.

---

### Dictionary Comprehension

Dictionary comprehension är en kortfattad syntax för att skapa ordböcker baserat på befintliga ordböcker eller itererbara objekt.

- **Exempel**:
  ```python
  square_dict = {x: x**2 for x in range(5)}
  print(square_dict)  # Output: {0: 0, 1: 1, 2: 4, 3: 9, 4: 16}
  ```

  Här skapar vi en ordbok där varje nyckel är ett tal och värdet är kvadraten av det talet.

---

### Statement

En statement är en kodrad som utför en specifik uppgift.

- **Exempel**:
  ```python
  x = 10
  ```

  Här är `x = 10` en statement som tilldelar värdet `10` till variabeln `x`.

---

### Argument

Argument är värden som skickas till en funktion när den anropas.

- **Exempel**:
  ```python
  def greet(name):
      print(f"Hello, {name}")

  greet('Alice')  # Output: Hello, Alice
  ```

  Här är `'Alice'` ett argument som skickas till funktionen `greet`.

---

### Boolean

Boolean är en datatyp som kan ha värdena `True` eller `False`, och används för att representera logiska värden.

- **Exempel**:
  ```python
  is_active = True
  if is_active:
      print("Active")  # Output: Active
  ```

  Här är `is_active` en boolean som används för att kontrollera ett villkor.

---

### Class

En klass är en mall för att skapa objekt. Den definierar attribut och metoder som objektet kommer att ha.

- **Exempel**:
  ```python
  class Person:
      def __init__(self, name, age):
          self.name = name
          self.age = age

  p = Person('Alice', 30)
  print(p.name)  # Output: Alice
  ```

  Här är `Person` en klass med attributen `name` och `age`, och vi skapar en instans av `Person` med namnet `'Alice'` och åldern `30`.

---

### Conditional

En conditional är en kodstruktur som utför olika kod beroende på om ett villkor är sant eller falskt.

- **Exempel**:
  ```python
  x = 10
  if x > 5:
      print("x is greater than 5")  # Output: x is greater than 5
  ```

  Här kontrollerar vi om `x` är större än `5` och skriver ut ett meddelande om det är sant.

---

### Exception

En exception är ett fel som uppstår under körning av ett program. Undantag kan hanteras med `try` och `except`.

- **Exempel**:
  ```python
  try:
      result = 10 / 0
  except ZeroDivisionError:
      print("Cannot divide by zero")  # Output: Cannot divide by zero
  ```

  Här försöker vi dela med noll, vilket orsakar ett undantag som vi fångar och hanterar.

---

### Function

En funktion är en block av kod som utför en specifik uppgift och kan återanvändas. Funktionen kan ta emot argument och returnera ett värde.

- **Exempel**:
  ```python
  def add(a, b):
      return a + b

  result = add(5, 3)
  print(result)  # Output: 8
  ```

  Här definierar vi en funktion `add` som lägger till två tal och returnerar resultatet.

---

### Keyword Arguments

Keyword arguments är argument som ges till en funktion genom att ange namnet på parametern. Detta gör att du kan ge argumenten i valfri ordning.

- **Exempel**:
  ```python
  def greet(name, greeting):
      print(f"{greeting}, {name}")

  greet(name='Alice', greeting='Hello')  # Output: Hello, Alice
  ```

  Här använder vi nyckelargument för att ange `name` och `greeting` i funktionen `greet`.

---

### Module

En modul är en fil med Python-kod som kan importeras och användas i andra Python-program.

- **Exempel**:
  ```python
  import math
  print(math.sqrt(16))  # Output: 4.0
  ```

  Här importerar vi `math`-modulen och använder dess `sqrt`-funktion för att beräkna kvadratroten av `16`.

---

### Object

Ett objekt är en instans av en klass. Det har attribut och metoder som definieras av klassen.

- **Exempel**:
  ```python
  class Person:
      def __init__(self, name):
          self.name = name

  p = Person('Alice')
  print(p.name)  # Output: Alice
  ```

  Här är `p` ett objekt av klassen `Person` med attributet `name` satt till `'Alice'`.

---

### Return Statement

En return statement är en instruktion i en funktion som skickar tillbaka ett värde till den plats där funktionen anropades.

- **Exempel**:
  ```python
  def square(x):
      return x * x

  result = square(4)
  print(result)  # Output: 16
  ```

  Här returnerar funktionen `
